```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Travel.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')
```

```python
df['MaritalStatus'].value_counts()
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace('Single','Unmarried')
```

```python
df.head()
```

```python
##checking missig values
df.isnull().sum()
```

```python
feature_with_na = [features for features in df.columns if df[features].isnull().sum()>=1]
for feature in feature_with_na:
    print(feature, np.round(df[feature].isnull().mean()*100,5), '% missing values')
```

```python
df.Age.fillna(df.Age.median(), inplace=True)
df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace=True)
df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace=True)
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace=True)
df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace=True)
df.NumberOfFollowups.fillna(df.NumberOfFollowups.median(), inplace=True)
df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```

```python
df.drop('CustomerID', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['totalvisits'] = df['NumberOfChildrenVisiting'] + df['NumberOfPersonVisiting']
```

```python
df.drop(columns=['NumberOfChildrenVisiting','NumberOfPersonVisiting'], inplace=True)
```

```python
##train test split
from sklearn.model_selection import train_test_split
x = df.drop('ProdTaken', axis=1)
y = df['ProdTaken']
```

```python
x.head()
```

```python
y.value_counts()    
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42,stratify=y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape    
```

```python
# create column tranformer with 3 types of tranformers
cat_features = x.select_dtypes(include='object').columns 
num_features = x.select_dtypes(exclude='object').columns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler 

numeric_transformer = StandardScaler()
oh_tranformer = OneHotEncoder(drop = 'first')

preprocessor = ColumnTransformer(
    [
        ("OnehotEncoder", oh_tranformer, cat_features),
        ("StandardScaler", numeric_transformer, num_features)
    ]
)

```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
x_train
```

```python
pd.DataFrame(x_train)
```

```python
y.value_counts()
```

```python
##apply random forest classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

```

```python
model = {
    "random forest": RandomForestClassifier()

}

for m in model:
    print(f"***********{m}************")
    model[m].fit(x_train, y_train)
    y_pred = model[m].predict(x_test)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
#training set performance
y_train_pred = model[m].predict(x_train)
model_train_acc = accuracy_score(y_train, y_train_pred)
print("Training Set Accuracy:", model_train_acc)
model_train_f1 = classification_report(y_train, y_train_pred)
print("Training Set Classification Report:\n", model_train_f1)
```

```python
#test set performance
y_test_pred = model[m].predict(x_test)
model_test_acc = accuracy_score(y_test, y_test_pred)
print("Test Set Accuracy:", model_test_acc)
model_test_f1 = classification_report(y_test, y_test_pred)
print("Test Set Classification Report:\n", model_test_f1)
```

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 5, 7,8]
}
```

```python
randomcv_models = [
    ("RF", RandomForestClassifier(), param_grid)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV
for name, model, params in randomcv_models:
    print(f"***********Tuning {name}************")
    random_search = RandomizedSearchCV(estimator=model, param_distributions=params, n_iter=10, cv=3, n_jobs=-1, verbose=2, random_state=42)
    random_search.fit(x_train, y_train)
    print(f"Best Parameters for {name}: {random_search.best_params_}")
    best_model = random_search.best_estimator_
    y_pred = best_model.predict(x_test)
    print("Accuracy after Hyperparameter Tuning:", accuracy_score(y_test, y_pred))
    print("Classification Report after Hyperparameter Tuning:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix after Hyperparameter Tuning:\n", confusion_matrix(y_test, y_pred))
```
